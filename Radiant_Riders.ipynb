```python
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd # Import pandas

# Define Hyperparameter Grid
param_grid = {
    'randomforest__n_estimators': [100, 300, 500],  # Prefix with pipeline step name
    'randomforest__max_depth': [5, 10, None],
    'randomforest__min_samples_split': [2, 5, 10],
    'randomforest__min_samples_leaf': [1, 2, 4]
}

# Define K-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store MSE scores for each fold
nested_mse_scores = []
best_params_list = []

# Loop through folds manually for better control
for train_idx, val_idx in kf.split(X_train):
    # Convert NumPy arrays back to pandas DataFrames for iloc to work
    X_fold_train = pd.DataFrame(X_train).iloc[train_idx]
    X_fold_val = pd.DataFrame(X_train).iloc[val_idx]
    y_fold_train = pd.Series(y_train).iloc[train_idx]
    y_fold_val = pd.Series(y_train).iloc[val_idx]

    # Convert back to numpy arrays after using iloc
    X_fold_train = X_fold_train.values
    X_fold_val = X_fold_val.values
    y_fold_train = y_fold_train.values
    y_fold_val = y_fold_val.values

    # Define Pipeline with Feature Scaling
    pipeline = Pipeline([
        ('scaler', MinMaxScaler()),
        ('randomforest', RandomForestRegressor(random_state=42))
    ])

    # Perform Hyperparameter Search
    grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='neg_mean_squared_error')
    grid_search.fit(X_fold_train, y_fold_train)

    # Get Best Model and Evaluate on Validation Set
    best_model = grid_search.best_estimator_
    mse = -grid_search.score(X_fold_val, y_fold_val)

    # Store Results
    nested_mse_scores.append(mse)
    best_params_list.append(grid_search.best_params_)

# Print Results
print(f"\nNested CV MSE: {np.mean(nested_mse_scores):.4f} ± {np.std(nested_mse_scores):.4f}")
print("\nBest Parameters per Fold:")
for i, params in enumerate(best_params_list):
    print(f"Fold {i+1}: {params}")
```
